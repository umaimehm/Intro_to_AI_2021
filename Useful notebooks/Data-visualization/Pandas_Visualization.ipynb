## Data Visualization 
It is the graphical representation of data in order to interactively convey insights to stakeholders. 
Python comes with data visualization libraries that helps us to visualize data efficiently.
The python libraries we are going to cover are:
1. Pandas Visualization
2. Matplotlib
3. Seaborn


### Pandas Visualization : easy to use interface, built on matplotlib.
First and Foremost is Importing Dataset. 
Here we will be using<b> Iris.csv</b> dataset.

```python
#import pandas library
import pandas as pd
```

```python
# Load the Iris.csv dataset using read_csv() function of Pandas dataframe.
iris = pd.read_csv('iris.csv',names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
```

```python
#Printing the rows 
iris.head()
```

### Scatter Plot 
Function used : <b>dataset_name.plot.scatter(name_of_x_col,name_of_y_col,title)</b>

```python
iris.plot.scatter(x='sepal_length', y='sepal_width', title='Iris Dataset')
```

### Line Chart
Function used : <b>dataframe.plot.line(title_name)</b>

```python
iris.drop(['class'], axis=1).plot.line(title='Iris Dataset')
```

### Histogram 
Function used : <b>dataframe.plot.hist()</b>

```python
#Histogram displaying only one feature
iris['sepal_length'].plot.hist()
```

```python
#Histogram displaying seperate plot for each feature
iris.plot.hist(subplots=True, layout=(2,2), figsize=(10, 10), bins=20)
```

### Bar Chart
Method used : <br>
<b>plot.bar()</b> This will create a vertical bar chart.<br>
<b>plot.barh()</b>. This will create a horizontal bar chart.
<br>
Before calling the above method, we need to count the occurences using <b>value.count()</b> method and sort the occurence in descending order using <b>sort_index()</b>method.

```python
iris['petal_width'].value_counts().sort_index().plot.bar()
```

```python
iris['petal_width'].value_counts().sort_index().plot.barh()
```

```python
iris.groupby("class").petal_width.mean().sort_values(ascending=False)[:5].plot.bar()
```
